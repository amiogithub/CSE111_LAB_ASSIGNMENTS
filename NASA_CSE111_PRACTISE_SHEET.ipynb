```python
#Task:02 (Maam problem: int object is not callable)   #Solved

class myList:
  def __init__(self,*args):
    self.add = 0    #Declired it as instance variable otherwise for l2 object there is no instance variable add
    self.a=[]
    for i in args:
      self.a.append(i)
  def sum(self):
    self.add = 0    #changed the variable name from sum to add because the method name was also sum. When you call l1.sum() it was actually calling the instance variable instead of the method
    for i in self.a:
      self.add+=i
    print('Sum:',self.add)

  def merge(self,*args):
    for i in args:
      self.a.append(i)

  def average(self):
    if self.add == 0:
      print('Average: 0')
    else:
      print('Average:',self.add/len(self.a))
           

l1 =  myList(2,3,4,5,6) 
l1.sum()
l1.merge(4,5,9)
l1.sum()
l1.average()
print('-----------------------------')
l2 =  myList()
l2.average()
l2.merge(1,2,4,8)
l2.sum()
```

```python
#Task:16(Complete)   
import math

class Vector2D:
  def __init__(self, Xcomponent, Ycomponent, vec_type = 'Default'):
    self.Xcomponent = Xcomponent
    self.Ycomponent = Ycomponent
    self.vec_type = vec_type
    
  def __str__(self):
    return str(self.Xcomponent)+"i + "+str(self.Ycomponent)+ "j"


class Vector3D(Vector2D):
  def __init__(self,Xcomponent, Ycomponent,Zcomponent, vec_type):
    super().__init__(Xcomponent, Ycomponent, vec_type)
    self.Zcomponent=Zcomponent

    # self.vector=f'{self.Xcomponent}i{self.Ycomponent}j{self.Zcomponent}k'

  def __str__(self):
    return f'{super().__str__()} {self.Zcomponent}k'

  def calculate_length(self):
    self.length=math.sqrt(self.Xcomponent**2+self.Ycomponent**2+self.Zcomponent**2) 
    return self.length  

  def calculate_unit_vector(self):
    self.unit=f'{self.Xcomponent/self.length}i + {self.Ycomponent/self.length}j + {self.Zcomponent/self.length}k'
    return self.unit

  def __add__(self,other):
    if self.vec_type==other.vec_type:
      return f'{self.Xcomponent+other.Xcomponent}i + {self.Ycomponent+other.Ycomponent}j + {self.Zcomponent+other.Zcomponent}'
    else:
      return f'Different type of vectors cannot be added'
  def __sub__(self,other):
        if self.vec_type==other.vec_type:
          return f'{self.Xcomponent-other.Xcomponent}i + {self.Ycomponent-other.Ycomponent}j + {self.Zcomponent-other.Zcomponent}k'
        else:
         return f'Different type of vectors cannot be subtracted'  

  def __mul__(self,other):
    self.dot=f'{self.Xcomponent*other.Xcomponent+self.Ycomponent*other.Ycomponent+self.Zcomponent*other.Zcomponent}'
    return self.dot


  def is_ortho(self,other):
    a= self.Xcomponent*other.Xcomponent+self.Ycomponent*other.Ycomponent+self.Zcomponent*other.Zcomponent   #Removed the f string 
    if a==0:
      return True
    else:
      return False


force1 = Vector3D(1, 3, 5, 'force')
print('---------------------------------------')
print(force1)
print('---------------------------------------')
print('length of force1 vector is: ', force1.calculate_length())
force1_unit_vector = force1.calculate_unit_vector()
print('unit vector of force1 vector is: ', force1_unit_vector)
print('---------------------------------------')
displacement1  = Vector3D(5, -5, 2, 'displacement')
print(displacement1)
print('----------------------------------------')
summ = force1 + displacement1
print('result of addition:', summ)
print('----------------------------------------')
force2 = Vector3D(7, 3, -2, 'force')
summ = force1 + force2
print('result of addition:', summ)
print('----------------------------------------')
diff = force1 - force2
print('result of subtraction:', diff)
print('----------------------------------------')
work = force1 * displacement1
print('work done by force1 and displacement1 is:', work)

print('-----------------------------------------------')
print('force1 and displacement1 are orthogonal:', force1.is_ortho(displacement1))
print('force2 and displacement1 are orthogonal:', force2.is_ortho(displacement1))
```

```python
#SUMMER 2022
#QUESTION 01

class Monster:
  monsterCount=0
  def __init__(self,name,power,alive=True):
    self.name=name
    self.power=power
    self.alive=alive
    Monster.monsterCount+=1

  def get_details(self):
    return f'Name:{self.name}\nPower:{self.power}\nAlive:{self.alive}' 

  def attack(self,*args):
    if len(args)==0:
      print('No monster to attack')
    else:
      for i in args:
       if self.alive==True:
        if i.alive==True:
          if self.power>i.power:
            print(f'Attack successful.{self.name} defeated {i.name}')
            i.alive=False
            Monster.monsterCount-=1
          else:
            print(f'Attack unsucessful. {self.name} was defeated by {i.name}')
            Monster.monsterCount-=1
        else:
          print(f"Cannot attack {i.name}. It's not alive")
      else:
        print(f'{self.name} is not alive to attack')

monster1 = Monster('Godzilla', 40)
monster2 = Monster('Hydra', 30)
monster3 = Monster('KingKong', 50)
print(f"Number of monsters alive:{Monster.monsterCount}")
print('1--------------------------------')
print(monster1.get_details())
print('2--------------------------------')
print(monster2.get_details())
print('3--------------------------------')
print(monster3.get_details())
print('4--------------------------------')
monster1.attack()
print('5--------------------------------')
monster1.attack(monster2)
print('6--------------------------------')
monster1.attack(monster2, monster3)
print('7--------------------------------')
print(f"Number of monsters alive:{Monster.monsterCount}")
print('8--------------------------------')
print(monster2.get_details())
print('9--------------------------------')
monster2.attack(monster1)
```
